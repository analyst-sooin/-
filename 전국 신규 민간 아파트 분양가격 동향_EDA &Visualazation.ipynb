## 규모구분 컬럼 전처리
규모구분 컬럼은 전용면적에 대한 내용이 있는 컬럼이다. 전용면적이라는 문구가 공통적으로 나와 있고, 규모구분보다 전용면적이 더 직관적이기 때문에,
컬럼 이름을 전용면적으로 바꿔주고 데이터들에서 전용면적이라는 문구와 초과, 이하등의 문자들을 제거하고 간결하게 보겠다.

```python
df_last['규모구분'].unique()
```

```python
df_last['전용면적'] = df_last['규모구분'].str.replace('전용면적','')
df_last['전용면적'] = df_last['전용면적'].str.replace('초과 ','~')
df_last['전용면적'] = df_last['전용면적'].str.replace('초과','~')
df_last['전용면적'] = df_last['전용면적'].str.replace('이하','')
df_last.head()
```

### melt로 데이터 형태 바꿔주기
2015년 데이터와 이전데이터의 데이터 형태가 다르므로 데이터를 합쳐주기 위해서 형태를 변경해준다

```python
display(df_first.sample())
df_last.sample()
```

```python
df_first_melt = pd.melt(df_first, id_vars='지역',var_name='기간',value_name='평당분양가격')
df_first_melt
```

```python
df_first_melt.columns = ['지역명','기간','평당분양가격']
df_first_melt
```

df_first_melt의 기간 컬럼을 df_last의 형태와 같이 연도와 열을 나눠 준다.
split 함수로 연도와 월을 분리한 후 replace로 문자열을 제거해준뒤 숫자타입으로 바꿔준다. (새로운 함수 만들기)

```python
def parse_year(date):                             # parse_year로 문자열이 들어오면 year로 만들어주는 함수명을 만든다
    
    year = date.split('년')[0].replace('년','')    # '년'을 기준으로 문자열을 분리한 후 0번째 index를 선택하여 '년'문자열을 제거한다.
    year = int(year)                              # int 타입으로 바꿔준다
    
    return year
```

```python
date = '2013년12월'
date
```

```python
parse_year(date)
```

```python
def parse_month(date):
    
    month = date.split('년')[1].replace('월','')
    month = int(month)
    
    return month
```

```python
parse_month(date)
```

새로 만든 함수가 잘 작동하는 것을 테스트 한뒤, apply함수를 이용하여 기존 테이블에 적용하고 새로은 컬럼을 만들어준다.

```python
df_first_melt['연도'] = df_first_melt['기간'].apply(parse_year)
df_first_melt
```

```python
df_first_melt['월'] = df_first_melt['기간'].apply(parse_year)
df_first_melt
```

```python
display(df_first_melt.sample())
df_last.sample()
```

```python
cols = ['지역명','연도','월','평당분양가격']
```

```python
# 최근 데이터가 담긴 df_last 에는 전용면적이 있다
# 이전 데이터에는 전용면적이 없기 때문에 "전체"만 사용
df_last_prepare = df_last.loc[df_last['전용면적'] == '전체', cols].copy()
df_last_prepare
```

```python
# df_first_melt에서 공통된 컬럼만 가져온다
df_first_prepare = df_first_melt[cols].copy()
df_first_prepare
```

연도별로 데이터가 몇개씩 존재하는지 확인해준다.

*sort =False로 해주면 데이터 갯수로 정렬을 하지 않는다.

```python
df['연도'].value_counts(sort = False)
```

```python
# 연도를 인덱스로, 지역명을 컬럼으로 평당분양가격을 피봇테이블로 집계
r = pd.pivot_table(data = df, index = '연도',columns = '지역명',values = '평당분양가격')
```

```python
# 히트맵으로 표현하기

plt.figure(figsize=(20,7))
sns.heatmap(r, cmap='Blues',annot = True, fmt='.0f')
```

색이 진할 수록 값(평당분양가격이 높은것으로 나온다. 전체적으로 연도가 높아질수록 색이 진해지는 것을 볼 수 있다.

```python
r.T #Transepose로 행열 바꿔주기
```

```python
# Transepose한 데이터로 히트맵 표현하기

plt.figure(figsize=(20,7))
sns.heatmap(r.T, cmap='Blues',annot = True, fmt='.0f')
```
